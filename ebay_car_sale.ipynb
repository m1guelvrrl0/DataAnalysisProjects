# Exploratory Data Analysis

```python
autos = pd.read_csv("autos.csv", encoding='Latin-1')
autos
```

```python
autos.info()
```

A few columns seem to have multiple null values, especially 'notRepairedDamage', 'fuelType', 'model', and 'gearbox' in addition 'nrOfPictures' seems to be populated by mostly 0's, while most other columns are populated by strings, lastly columns names use CamelCase.

```python
#Converting CamelCase to snake_case
print(autos.columns)  
autos.columns = autos.columns.str.replace('([a-z0-9])([A-Z])', r'\1_\2').str.lower()
```

```python
autos.describe(include='all')
```

```python
#price and odometer need to clenaned for the $ and km respectively. 
autos["price"] = autos["price"].str.replace('$','').str.replace(',','').astype(float)
autos["odometer"] = autos["odometer"].str.replace('km','').str.replace(',','').astype(float)
autos.rename(columns={"odometer":"odometer_km"}, inplace=True)
print(autos[["price","odometer_km"]])
```

```python
#Look for potential outliers in price and odometer_km
print(autos.price.describe().apply(lambda x: format(x, 'f'))) #lambda to supress scientific notation convertion
print(autos.price.value_counts().head(10).sort_index(ascending=True))
print(autos.odometer_km.describe())
print(autos.odometer_km.value_counts().head(10).sort_index(ascending=True))
```

```python
# On price there seems to be a lot of 0's and the max is obviously not reliable
print(autos.price.describe(percentiles=[0.10, 0.90])) #Find the 10th percentile and 90th percentile 
autos = autos[(autos["price"] > 500 ) & (autos["price"] < 14000 )] #Remove outliers using the percentile values
```

```python
#Verify we removed outliers
print(autos.price.value_counts().head(10).sort_index(ascending=True))
```

```python
#Seems much better now
autos.describe(include='all')
```

### Brands and price

```python
#Making a dict of the mean price of the most common brands
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
mean_price = autos.groupby(autos["brand"])["price"].mean().round(2)
mean_price_dict = mean_price.to_dict()
mean_price_top = {common_brands: mean_price_dict[common_brands] for common_brands in common_brands}
print(mean_price.describe())
print(mean_price_top) 
```

There seems be considerable difference between the brands, with bmw, mercedes and audi being considerably higher priced than ford and opel.

```python
autos[autos["brand"].isin(common_brands)].groupby(autos["brand"])["price"].mean().round(2).plot(kind='bar') 
```

```python
autos[autos["brand"].isin(common_brands)].groupby(autos["brand"])["price"].std().sort_values()
```
